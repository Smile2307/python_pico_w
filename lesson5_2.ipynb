```python
#國文100,同時數學100,獎金1000
#國文100,或者數學100,獎金500
#其它獎金0

chinese = pyip.inputInt("請輸入國文分數:",min=0,max=100)
print(chinese)
math = pyip.inputInt("請輸入數學分數:",min=0,max=100)
print(math)
bonus = 0

#巢狀判斷
if chinese == 100:
    if math == 100:
        bonus = 1000
    else:
        bonus = 500
elif math == 100:
    bonus = 500

print(f"獎金:{bonus}")
```

```python
#國文100,同時數學100,獎金1000
#國文100,或者數學100,獎金500
#其它獎金0

chinese = pyip.inputInt("請輸入國文分數:",min=0,max=100)
print(chinese)
math = pyip.inputInt("請輸入數學分數:",min=0,max=100)
print(math)
bonus = 0

#邏輯運算子
if chinese == 100 and math == 100:
    bonus = 1000
elif chinese == 100 or math == 100:
    bonus = 500

print(f"獎金:{bonus}")
```
